# <center> TRANSCRIPTOME </center>

<b>Transcription of a gene expression involves a particular segment of DNA being copied into RNA by the enzyme RNA polymerase.</b>
The transcriptome is the set of all RNA molecules, including mRNA, rRNA, tRNA, and other non-coding RNA produced in one or a population of cells. It differs from the exome in that it includes only those RNA molecules found in a specified cell population, and usually includes the amount or concentration of each RNA molecule in addition to the molecular identities.

<center> <h2 style="color:purple;"> Generic Feature Format (GFF) </h2></center> 

### GFF3 parsing
Data storage in GFF is not record oriented unlike other biological data formats. Rather, the data is stored in a line-oriented hierarchical format with support for nesting feature lists. Also GFF only stores the biological features, not the entire primary sequence of the species. This poses some challenges while trying to parse it to browse annotations. 

Lets start by importing the GFF libraries:

```python
import sys, pprint
import Bio
from BCBio.GFF import GFFParser, GFFExaminer 
```

### How to parse the GFF file
Due to the unordered data storage structure, it is difficult to find any identifiers running through the gene trees. Hence, to obtain a relevant record: 
- one has to do multiple passes on the file and examine it to determine available annotations.
- As the gene files may have many annotations, it will hog the memory when you load the large file. Hence, parsing must be restricted to "<b> biological features of interest</b>" and our program must only "<b>iterate over some portions of the file</b>".
- Sequences must be parsed separately, from an associated FASTA file.

##### Limited feature parsing
Let us parse the '[dana-scaffold_0-r1.3.gff](https://gitlab.com/aleph-omega/flybase/blob/master/dana-gff_all_scaffolds-r1.3/dana-scaffold_0-r1.3.gff)' for limited biological features

```python
in_file = "/home/mom/Aleph-Omega/flybase/dana-gff_all_scaffolds-r1.3/dana-scaffold_0-r1.3.gff"
# in_file_s1 = "/home/mom/Aleph-Omega/flybase/dana-gff_all_scaffolds-r1.3/dana-scaffold_1-r1.3.gff"
parser = GFFParser()
limit_info = dict(
        gff_id = ["Chr1"],
        gff_type = ["protein", "gene", "mRNA", "CDS", "exon"],
        )
examiner = GFFExaminer()
in_handle = open(in_file)
for rec in examiner.parse(in_handle, limit_info = limit_info):
    print(rec.features[0])
in_handle.close()
```

<center> <h2 style="color:purple;"> Retrieving a subset of the chromosome1 sequence features to parse.</h2></center> 

Lets parse the file "[dana-all-no-analysis-r1.3.gff](https://gitlab.com/aleph-omega/flybase/blob/master/dana-all-no-analysis-r1.3.gff)" filtered from the flybase DB, to see how the data is tagged for various DNA features. From the parsed output, the various tags used are:

* gff_id => Description of a record identifier that refers to a chromosome or other scaffold sequence.
* gff_source => is the source description from the second column of the GFF file, which specifies how a feature was generated.
* gff_source_type => All combinations of sources and types in the file.
* gff_type => The type of the feature, pulled from the 3rd column of the GFF file.

The available_limits function in the examiner gives you a high level summary of these feature attributes, along with counts for the number of times they appear in the file:

```python
import sys, pprint
import Bio
from BCBio.GFF import GFFParser, GFFExaminer 

in_file = '/home/mom/Aleph-Omega/flybase/dana-all-no-analysis-r1.3.gff'
parser = GFFParser()
limit_info = dict(
        gff_id = ["Chr1"],
        gff_type = ["protein", "gene", "mRNA", "CDS", "exon"],
        )
examiner = GFFExaminer()
in_handle = open(in_file)
pprint.pprint(examiner.available_limits(in_handle))
in_handle.close()
```
